In this example, we will calculate term frequenct inverse document frequency <B>(TFIDF).</B>
We will use<b> sklearn </b> to calculate TFIDF.
We will also use a simple dataset. The dataset contains simple 5 documents just to understand the usage of TF and TFIDF.


```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

import pandas as pd
 
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer

from google.colab import files
uploaded = files.upload()

#module for digits - we use this to remove digits
from string import digits

docs = []
for file in uploaded.keys():

  with open(file, "r") as file:
    FileContent = file.read()
    print("File Name: ",file.name)
    #file content
    #print("File Content: ",FileContent)
    remove_digits = str.maketrans('', '', digits) 
    FileContent = FileContent.translate(remove_digits) 
    FileContent = FileContent.strip()
    FileContent = FileContent.lower()
    docs.append(FileContent)

# this is a very toy example, do not try this at home unless you want to understand the usage differences
#docs = ["the house had a tiny little mouse",
#        "the cat saw the mouse",
#        "the mouse ran away from the house",
#        "the cat finally ate the mouse",
#        "the end of the mouse story"
#        ]
 
```

To start using TfidfTransformer you will first have to create a CountVectorizer to count the number of words (term frequency), limit your vocabulary size, apply stop words and etc.

```
from sklearn.feature_extraction.text import CountVectorizer

# Create CountVectorizer object
vectorizer = CountVectorizer()

# Generate matrix of word vectors
bow_matrix = vectorizer.fit_transform(docs)

# Print the shape of bow_matrix
print(bow_matrix.shape)
```

```
# Convert bow_matrix into a DataFrame
bow_df = pd.DataFrame(bow_matrix.toarray())

# Map the column names to vocabulary
bow_df.columns = vectorizer.get_feature_names()

# Print bow_df
bow_df
```

```
stop_words = set(stopwords.words('english'))
#instantiate CountVectorizer()
#cv=CountVectorizer(ngram_range=(1,2),stop_words=stop_words)
cv=CountVectorizer(stop_words=stop_words)

 
# this steps generates word counts for the words in your docs
word_count_vector=cv.fit_transform(docs)

#print(word_count_vector)
 
```

let’s check the shape. We should have 15 rows (15 docs). (15, X). X will represent the number of unique word in our dataset.

```
word_count_vector.shape
```

```
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count_vector)
 
```

We are going to compute the IDF values by calling tfidf_transformer.fit(word_count_vector) on the word counts we computed earlier.

```
# print idf values
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=cv.get_feature_names(),columns=["idf_weights"])
 
# sort ascending
df_idf.sort_values(by=['idf_weights'])
```

```
# count matrix
count_vector=cv.transform(docs)
 
# tf-idf scores
tf_idf_vector=tfidf_transformer.transform(count_vector)
```

Once you have the IDF values, you can now compute the tf-idf scores for any document or set of documents. Let’s compute tf-idf scores for the 5 documents in our collection.Internally this is computing the <b>tf * idf </b>  multiplication where your term frequency is weighted by its IDF values.

<br> <b>The more common the word across documents, the lower its score and the more unique a word is to our first document (e.g. ‘good’ and ‘know’) the higher the score. </b>

```
feature_names = cv.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=tf_idf_vector[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer 
 
# settings that you use for count vectorizer will go here
tfidf_vectorizer=TfidfVectorizer(use_idf=True, stop_words=stop_words, ngram_range=(1,2))
 
# just send in all your docs here
tfidf_vectorizer_vectors=tfidf_vectorizer.fit_transform(docs)
```

```
# get the first vector out (for the first document)
first_vector_tfidfvectorizer=tfidf_vectorizer_vectors[1]
 
# place tf-idf values in a pandas data frame
df = pd.DataFrame(first_vector_tfidfvectorizer.T.todense(), index=tfidf_vectorizer.get_feature_names(), columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
 
```
