## Step 2: Text Preprocessing - Clean the data

#### Start the text preprocessing

The text preprocessing includes:
1. convert tweet to lower case
2. remove non-english tweets
3. remove url
4. remove username
5. replace emojis to words
6. remove punctuations
7. remove stopwords
8. lemmatization

```python
import string
import re
import emoji
import nltk
from nltk.corpus import stopwords
from langdetect import detect 
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

# Convert tweet to lower case
tweet_text = tweet_text.str.lower()

# Function to detect language
def detect_language(text,target_lang):
    result_lang = detect(text)
    if result_lang == target_lang:
        return text 
    else:
        return None

# Remove url 
def remove_url(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# Remove username as they don't contain meaningful information
def remove_username(text):
  return re.sub("@[A-Za-z0-9]+","", text)

# Replace emojis to words
def emojis_to_words(text):
  return emoji.demojize(text)

# Remove punctuations 
# Function to remove punctuations 
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

# Remove stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

# Lemmatization
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])


tweet_text= tweet_text.apply(lambda x:detect_language(x,'en'))
# Remove non-english tweets
tweet_text.dropna(inplace=True)
print('Size of tweet text after removing non-english tweets: ',str(len(tweet_text)))
tweet_text= tweet_text.apply(lambda x:remove_url(x))
tweet_text= tweet_text.apply(lambda x:remove_username(x))
tweet_text= tweet_text.apply(lambda x:emojis_to_words(x))
tweet_text= tweet_text.apply(lambda x:remove_punctuation(x))
tweet_text= tweet_text.apply(lambda x:remove_stopwords(x))
tweet_text= tweet_text.apply(lambda x:lemmatize_words(x))
tweet_text
```

## Step 3: Visualize the data with wordcloud

####Combine our tokens/words back together to create a big string and count how many words we have



```python
text = " ".join(tweet for tweet in tweet_text)
# checking how many words we have
print("Words: {}".format(len(text)))
```

#### Import the wordcloud library and add stopwords

```python
# Import the wordcloud library
from wordcloud import WordCloud
# Instantiate a new wordcloud.
wordcloud = WordCloud(random_state = 8,
        normalize_plurals = False,
        width = 600, height= 300,
        max_words = 300,
        stopwords = ['concert','also','want','wan','wanna','get','loudlycryingface','go','going','concerts','tickets','im','ticket','even'])


# Apply the wordcloud to the text.
wordcloud.generate(text)
```

```python
import matplotlib.pyplot as plt
# Create a figure
fig, ax = plt.subplots(1,1, figsize = (10,8))
# Add interpolation = bilinear to smooth things out
plt.imshow(wordcloud, interpolation='bilinear')
# Remove the axis
plt.axis("off")
```
